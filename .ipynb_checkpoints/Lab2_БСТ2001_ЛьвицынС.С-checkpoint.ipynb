## Бинарный поиск

```python
import random
import timeit

N=10
arr=[]
for i in range(N):
    arr.append(random.randint(1, 10))
print(arr)
def binSearch(lys, val):
    arr.sort()
    first = 0
    last = len(lys)-1
    index = -1
    while (first <= last) and (index == -1):
        mid = (first+last)//2
        if lys[mid] == val:
            index = mid
        else:
            if val<lys[mid]:
                last = mid -1
            else:
                first = mid +1
    return index

answerAdd = input('Хотите ли вы добавить элемент в исходный массив? Введите "да" или "нет". Ваш ответ:  ')
if answerAdd == "да":
    addElement = int(input("Введите новый элемент = "))
    arr.append(addElement)
    print(arr)
print()

answerDel = input('Хотите ли вы удалить элемент из исходного массив? Введите "да" или "нет". Ваш ответ:  ')
if answerDel == "да":
    delElement = int(input("Введите индекс элемента, который необходимо удалить = "))
    arr.pop(delElement)
    print(arr)
print()

num=int(input("Введите число, которое хотите найти: ",))
if num>=1 and num<=10:
    start_time=timeit.default_timer()
    if(binSearch(arr, num) != -1):
        print("Данное число найдено!")
    else:
        print("Такого числа нет!")
    print("Работа выполнена за " + str(timeit.default_timer()-start_time)+" сек")
else:
    print("Введено неправильное число")
```

## Бинарное дерево

```python
import random
import timeit

N=10
arr=[]
for i in range(N):
    arr.append(random.randint(1, 10))
print(arr)

# поиск через бинарное дерево 
# Здесь происходит создание корневого узла
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Сортировка дерева
def inorder(root):
    if root is not None:
        # Обход левого поддерева
        inorder(root.left)
        # Обход корня
        print(str(root.data) + " ", end=' ')
        # Обход правого поддерева
        inorder(root.right)

# Вставка элемента
def insert(node, data):
    # Если дерево пустое, то вернуть новый узел
    if node is None:
        return Node(data)
    # Переход в левое/правое поддерево и вставка узла
    if data < node.data:
        node.left = insert(node.left, data)
    else:
        node.right = insert(node.right, data)
    return node

# Поиск inorder-преемника
def minValueNode(node):
    current = node
    # Inorder-преемник - это найденный крайний левый лист
    while(current.left is not None):
        current = current.left
    return current

# Удаление узла
def deleteNode(root, data):
    # Если дерево пустое, возвращаем
    if root is None:
        return root
    # Необходимо найти узел, который нужно удалить
    if data < root.data:
        root.left = deleteNode(root.left, data)
    elif(data > root.data):
        root.right = deleteNode(root.right, data)
    else:
        # Если у узла только один дочерний узел или вообще их нет
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        # В случае когда у узла два дочерних узла, помещаем центрированного преемника на место узла, который нужно удалить
        temp = minValueNode(root.right)
        root.data = temp.data
        # Удаляем inorder-преемниа
        root.right = deleteNode(root.right, temp.data)
    return root

root = None
for i in range(len(arr)):
    root = insert(root, arr[i])

print("Поиск через бинарное дерево этого же массива: ", end=' ')
inorder(root)
print()

answerAdd = input('Хотите ли вы добавить элемент в исходный массив? Введите "да" или "нет". Ваш ответ:  ')
if answerAdd == "да":
    addElement = int(input("Введите новый элемент = "))
    arr.append(addElement)
    print(arr)
print()

answerDel = input('Хотите ли вы удалить элемент из исходного массив? Введите "да" или "нет". Ваш ответ:  ')
if answerDel == "да":
    delElement = int(input("Введите индекс элемента, который необходимо удалить = "))
    arr.pop(delElement)
    print(arr)
print()
```

## Фибоначчиев

```python
import random
import timeit

N=10
arr=[]
for i in range(N):
    arr.append(random.randint(1, 10))
print(arr)
def fibonacci_search(lst, target):
    arr.sort()
    size = len(lst)    
    start = -1
    f0 = 0
    f1 = 1
    f2 = 1
    while(f2 < size):
        f0 = f1
        f1 = f2
        f2 = f1 + f0     
    while(f2 > 1):
        index = min(start + f0, size - 1)
        if lst[index] < target:
            f2 = f1
            f1 = f0
            f0 = f2 - f1
            start = index
        elif lst[index] > target:
            f2 = f0
            f1 = f1 - f0
            f0 = f2 - f1
        else:
            return index
    if (f1) and (lst[size - 1] == target):
        return size - 1
    return None

answerAdd = input('Хотите ли вы добавить элемент в исходный массив? Введите "да" или "нет". Ваш ответ:  ')
if answerAdd == "да":
    addElement = int(input("Введите новый элемент = "))
    arr.append(addElement)
    print(arr)
print()

answerDel = input('Хотите ли вы удалить элемент из исходного массив? Введите "да" или "нет". Ваш ответ:  ')
if answerDel == "да":
    delElement = int(input("Введите индекс элемента, который необходимо удалить = "))
    arr.pop(delElement)
    print(arr)
print()

num=int(input("Введите число, которое хотите найти: ",))
if num>=1 and num<=10:
    start_time=timeit.default_timer()
    if(fibonacci_search(arr, num) != None):
        print("Данное число найдено!")
    else:
        print("Такого числа нет!")
    print("Работа выполнена за " + str(timeit.default_timer()-start_time)+" сек")
else:
    print("Введено неправильное число")
```

## Интерполяционный

```python
import random
import timeit

N=10
arr=[]
for i in range(N):
    arr.append(random.randint(1, 10))
print(arr)
def IPSearch(lys, val):
    arr.sort()
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            return index
        if lys[index] < val:
            low = index + 1;
        else:
            high = index - 1;
    return -1

answerAdd = input('Хотите ли вы добавить элемент в исходный массив? Введите "да" или "нет". Ваш ответ:  ')
if answerAdd == "да":
    addElement = int(input("Введите новый элемент = "))
    arr.append(addElement)
    print(arr)
print()

answerDel = input('Хотите ли вы удалить элемент из исходного массив? Введите "да" или "нет". Ваш ответ:  ')
if answerDel == "да":
    delElement = int(input("Введите индекс элемента, который необходимо удалить = "))
    arr.pop(delElement)
    print(arr)
print()

num=int(input("Введите число, которое хотите найти: ",))
if num>=1 and num<=10:
    start_time=timeit.default_timer()
    if(IPSearch(arr, num) != -1):
        print("Данное число найдено!")
    else:
        print("Такого числа нет!")
    print("Работа выполнена за " + str(timeit.default_timer()-start_time)+" сек")
else:
    print("Введено неправильное число")
```

## Простое рехэширование

```python
# Создать класс хэш-таблицы
class Hash_Table:
    size = int(input("Введите, пожалуйста, размер хэш-таблицы: "))
    def __init__(self):
        self.data = [None] * self.size
    # Здесь получение содержимого баккета
    def __getitem__(self, key):
        h = self.get_Hash(key)
        try:
            while self.data[h]:
                if self.data[h] and self.data[h].get(key):
                    return self.data[h]
                h = self.get_Rehash(h)
        except IndexError:
            return
    # Здесь заполняется пустой баккет
    def __setitem__(self, key, value):
        h = self.get_Hash(key)
        if self.data[h] is None:
            self.data[h] = {key: value}
            return
        next_h = self.get_Rehash(h)
        try:
            while self.data[next_h] is not None:
                if self.data[next_h].get(key):
                    self.data[next_h].get(key)
                    break
                next_h = self.get_Rehash(next_h)
        except IndexError:
            raise
        self.data[next_h] = {key: value}
    # Сама хэш-функция
    def get_Hash(self, name):
        return len(name) % self.size
    # Это функция рехэширования
    def get_Rehash(self, oldhash):
        return oldhash + 1
# Заполнение хэш-таблицы двумя значениями
table = Hash_Table()
table[input('Введите ключ: ')] = input('Введите значение: ')
table[input('Введите ключ: ')] = input('Введите значение: ')
# Вывод одного из значений через ключ
print(table[input('Выберите значение по ключу: ')])
```

## Рехэширование с помощью псевдослучайных чисел

```python
class randHash:
    def __init__(self):
        self.data = [None]*16   
    def randNumbers(self,i):
        R=[5]
        if len(R) - 1 < i:
            R.append(self.randNumbers(i - 1))
        return (8255 * R[i] + 239) % 64   
    def _hash(self, key,i):
        return (hash(key)+self.randNumbers(i))%(2**4)
    def addItem_Hash(self,key):
        i=0
        while self.data[self._hash(key,i)] != None and i<len(self.data):
            i+=1
        self.data[self._hash(key,i)] = key
        return self.data   
    def searchHash(self,key):
        i=0
        while self.data[self._hash(key,i)] != None and i<len(self.data):
            if self.data[self._hash(key,i)] == key:
                return 1
            i+=1
        return 0
    def deleteItem_Hash(self,key):
        i=0
        while self.data[self._hash(key,i)] != None and i<len(self.data):
            if self.data[self._hash(key,i)] == key:
                self.data.pop(self._hash(key,i))
                return self.data
            i+=1
        print("Такого элемента нет")
Digital = randHash()
Digital.addItem_Hash("4")
Digital.addItem_Hash("11")
Digital.addItem_Hash("12")
Digital.searchHash("4")
Digital.deleteItem_Hash("11")
print(Digital.data)
```

## Метод цепочек

```python
class ChainMethod:
    def __init__(self):
        self.data = [None]*16
    def _hash(self,key):
        return hash(key)%(2**4)
    def addItem(self, index, key):
        if self.data[index] is None:
            self.data[index] = [key]
            print("элемент был добавлен")
            return 1
        else:
            for i, item in enumerate(self.data[index]):
                if item == key:
                    self.data[index][i] = key
                    print("элемент не был добавлен")
                    return 0
            self.data[index].append(key)
            print("элемент был добавлен")
            return 1
    def searchChain(self,index,item):
        if self.data[index] is not None:
            if self.data[index] == [item]:
                print("Элемент был найден")
                return 1
            else:
                for i, item_ in enumerate(self.data[index]):
                    if self.data[index][i] == item:
                        print("Элемент был найден")
                        return 1
        else:
            print("Элемент не найден")
            return 0
    def deleteItem_Chain(self,index,item):
        if self.data[index] is not None:
            if self.data[index] == [item]:
                self.data[index] = None
                print("Элемент был удален")
                return 1
            else:
                for i, item_ in enumerate(self.data[index]):
                    if self.data[index][i] == item:
                        self.data[index].pop(i)
                        print("Элемент был удален")
                        return 1
        else:
            return 0
tableChain = ChainMethod()
tableChain.addItem(tableChain._hash("Обед"),"Обед")
tableChain.addItem(tableChain._hash("Ужин"),"Ужин")
tableChain.addItem(tableChain._hash("Завтрак"),"Завтрак")
tableChain.searchChain(tableChain._hash("Завтрак"),"Завтрак")
tableChain.deleteItem_Chain(tableChain._hash("Обед"),"Обед")
print(tableChain.data)
```

### Задание 3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого.

```python
def solve_queen_problem(): 
    columns = set() 
    positive_diagonal = set() 
    negative_diagonal = set() 
    result = [] 
    board = [['. '] * 8 for _ in range(8)] 
  
    def backtrack(row: int): 
        if row == 8: 
            copy = [''.join(row) for row in board] 
            result.append(copy) 
            return 
  
        for column in range(8): 
            if column in columns or (row + column) in positive_diagonal or (row - column) in negative_diagonal: 
                continue 
            columns.add(column) 
            positive_diagonal.add(row + column) 
            negative_diagonal.add(row - column) 
            board[row][column] = 'Q ' 
            backtrack(row + 1) 
  
            columns.remove(column) 
            positive_diagonal.remove(row + column) 
            negative_diagonal.remove(row - column) 
            board[row][column] = '. ' 
  
    backtrack(0) 
    for x in result: 
        for y in x: 
            print(y) 
        print()

print(solve_queen_problem())
```
